#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

> Indeks: </br>
> Zawiera wiersze o następujących nazwach: mean, std, Q1, Q2, Q3, mode.

> Kolumny: lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str) </br>

Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE
index = [ 'mean', 'std', 'Q1', 'Q2', 'Q3', 'mode' ]
columns = [ '2006', '2007', '2008', '2009', '2010', '2011' ]
descriptive_parameters_2 = pd.DataFrame( index = index, columns = columns )

#kwantyle q1, q2, q3
rec_table_cumsum = recession_table_1.cumsum() / recession_table_1.sum()
cs = recession_table_1.cumsum()

tf1 = rec_table_cumsum < 0.25
tf2 = rec_table_cumsum < 0.5
tf3 = rec_table_cumsum < 0.75

for i in rec_table_cumsum.columns:
    xq1 = tf1.index[ tf1.loc[ :, i ] == False ][0]
    n4 = recession_table_1.loc[ :, i ].sum() / 4
    loc_sum = cs.shift(1).loc[xq1, i]
    nq1 = recession_table_1.loc[ xq1, i ]
    iq1 = xq1.right - xq1.left
    xq1 = xq1.left
    Q1 = xq1 + ( ( n4 - loc_sum ) / nq1 ) * iq1
    descriptive_parameters_2.loc[ 'Q1', i ] = Q1

    xq2 = tf2.index[ tf2.loc[ :, i ] == False ][0]
    n2 = recession_table_1.loc[ :, i ].sum() / 2
    loc_sum = cs.shift(1).loc[xq2, i]
    nq2 = recession_table_1.loc[ xq2, i ]
    iq2 = xq2.right - xq2.left
    xq2 = xq2.left
    Q2 = xq2 + ( ( n2 - loc_sum ) / nq2 ) * iq2
    descriptive_parameters_2.loc[ 'Q2', i ] = Q2
    
    xq3 = tf3.index[ tf3.loc[ :, i ] == False ][0]
    n4 = recession_table_1.loc[ :, i ].sum() / 4
    loc_sum = cs.shift(1).loc[xq3, i]
    nq3 = recession_table_1.loc[ xq3, i ]
    iq3 = xq3.right - xq3.left
    xq3 = xq3.left
    Q3 = xq3 + ( ( 3*n4 - loc_sum ) / nq3 ) * iq3
    descriptive_parameters_2.loc[ 'Q3', i ] = Q3
     
    #mean
    mean=0; std=0
    for index in recession_table_1[i].index:
        m=(((index.right+index.left)/2)*recession_table_1[i].loc[index])
        mean = mean + m
    
    mean = mean/sum(recession_table_1[i])
    
    #std
    for index in recession_table_1[i].index:
        std = std + (( ((index.right+index.left)/2) - mean)**2)*recession_table_1[i].loc[index]

    std = std/sum(recession_table_1[i])
    
    descriptive_parameters_2.loc[ 'mean', i ] = mean
    descriptive_parameters_2.loc[ 'std', i ] = np.sqrt(std)
    
    #moda
    max_index = int(np.argmax(recession_table_1[i]))
    indx = recession_table_1[i].index[max_index]
    
    table = recession_table_1[i].reset_index(drop=True)
    
    nd = table[max_index]
    nd_1 = table[max_index-1] 
    nd1 = table[max_index+1]
    hD = indx.length

    D = indx.left + ((nd-nd_1)/((nd-nd_1)+(nd-nd1)))*hD
    descriptive_parameters_2.loc[ 'mode', i ] = D
    
display( descriptive_parameters_2 )

```

```python
### Komórka testowa
# Testy wybranych wartości z tablicy
assert np.isclose(descriptive_parameters_2.loc["mean", "2006"], 5.606061)
assert np.isclose(descriptive_parameters_2.loc["std", "2007"], 4.184001)
assert np.isclose(descriptive_parameters_2.loc["Q1", "2008"], 0.916667)
assert np.isclose(descriptive_parameters_2.loc["Q2", "2009"], 0.220588)
assert np.isclose(descriptive_parameters_2.loc["Q3", "2010"], 7.604167)
assert np.isclose(descriptive_parameters_2.loc["mode", "2011"], 3.691589)
```
