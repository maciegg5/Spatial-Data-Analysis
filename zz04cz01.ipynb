### Zadanie 1: Podstawowe rozkłady teoretyczne wykorzystywane w statystyce matematycznej [15 pkt]
Przeanalizuj własności podstawowych teoretycznych rozkładów prawdopodobieństwa wykorzystywane w statystyce matematycznej - rozkładu normalnego, rozkładu t-Studenta, rozkładu chi-kwadrat i rozkładu F Snedecora.

#### a) Wizualizacja

Zwizualizuj funkcje gęstości prawdopodobieństwa następujących rozkładów teoretycznych:
- Wykres 1: $N(0, 1)$, $N(0, 0.75)$, $N(0, 1.25)$, $N(-2, 1)$, $N(2, 1)$
- Wykres 2: $N(0, 1)$, $t(1)$, $t(2)$, $t(5)$, $t(30)$
- Wykres 3: $\chi^2(1)$, $\chi^2(2)$, $\chi^2(3)$, $\chi^2(5)$, $\chi^2(8)$
- Wykres 4: $F(10, 1)$, $F(10, 2)$, $F(10, 5)$, $F(10, 10)$, $F(10, 20)$
- Wykres 5: $F(1, 10)$, $F(2, 10)$, $F(5, 10)$, $F(10, 10)$, $F(20, 10)$
- Wykres 6: $F(1, 1)$, $F(2, 2)$, $F(5, 5)$, $F(25, 25)$, $F(50, 50)$

gdzie:
- $N(\mu, \sigma)$ - rozkład normalny o wartości oczekiwanej $\mu$ i odchyleniu standardowym $\sigma$,
- $t(df)$ - rozkład t-Studenta o $df$ stopniach swobody,
- $\chi^2(df)$ - rozkład chi-kwadrat o $df$ stopniach swobody,
- $F(df1, df2)$ - rozkład F Snedecora o $df1$ i $df2$ stopniach swobody.

Zestaw wykresy na pojedynczej figurze (siatka wykresów 2x3). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw identyczne zakresy osi x i y wykresów 1, 2,
- ustaw identyczne zakresy osi x i y wykresów 4, 5, 6,
- dobierz zakresy osi x wszystkiech wykresów w taki sposób, aby dobrze widoczny był charakter znajdujących się na nich funkcji,
- zadbaj o to aby rozkład $N(0, 1)$ na obu wykresach był oznaczony w ten sam sposób (kolor, grubość, styl linii itd.),
- do każdego wykresów dodaj legendę opisującą widoczne na nim rozkłady i zadbaj o to, aby legenda nie zasłaniała innych elementów wykresu.

```python
fig, axs = plt.subplots(2,3, figsize=(15,10), facecolor="white")

x=np.linspace(-10,10,1000)

mu=[0,0,0,-2,2]
sigma=[1,0.75,1.25,1,1]

df_t=[1,2,5,30]
df_chi=[1,2,3,5,8]

df2_df1=[1,2,5,10,20]
df_6=[1,2,5,25,50]

for mu_val, sigma_val in zip(mu, sigma):
    axs[0, 0].plot(x, sp.stats.norm.pdf(x, mu_val, sigma_val), label=f'N({mu_val},{sigma_val})')

axs[0, 0].set_title('Rozkład normalny')
axs[0, 0].set_xlim(-7.5, 7.5)
axs[0, 0].set_ylim(0, 0.6)
axs[0, 0].legend()
   
axs[0, 1].plot(x, sp.stats.norm.pdf(x, 0, 1), label=f'N(0,1)')
for df in df_t:
    axs[0, 1].plot(x, sp.stats.t.pdf(x, df), label=f'T({df})')
axs[0, 1].set_title('Rozkład studenta')
axs[0, 1].set_xlim(-7.5, 7.5)
axs[0, 1].set_ylim(0, 0.6)
axs[0, 1].legend()

for chi in df_chi:
    axs[0, 2].plot(x, sp.stats.chi2.pdf(x, chi), label=f'Chi2({chi})')
axs[0, 2].set_title('Rozkład Chi2')
axs[0, 2].set_xlim(0, 10)
axs[0, 2].set_ylim(0, 0.8)
axs[0, 2].legend()

for df2 in df2_df1:
    axs[1, 0].plot(x, sp.stats.f.pdf(x, 10, df2), label=f'F(10,{df2})')

axs[1, 0].set_title('Rozkład Snedecora')
axs[1, 0].set_xlim(0, 6)
axs[1, 0].set_ylim(0, 1.5)
axs[1, 0].legend()

for df1 in df2_df1:
    axs[1, 1].plot(x, sp.stats.f.pdf(x, df1, 10), label=f'F({df1},10)')

axs[1, 1].set_title('Rozkład Snedecora')
axs[1, 1].set_xlim(0, 6)
axs[1, 1].set_ylim(0, 1.5)
axs[1, 1].legend()

for df in df_6:
    axs[1, 2].plot(x, sp.stats.f.pdf(x, df, df), label=f'F({df},{df})')

axs[1, 2].set_title('Rozkład Snedecora')
axs[1, 2].set_xlim(0, 6)
axs[1, 2].set_ylim(0, 1.5)
axs[1, 2].legend()

for i, ax in enumerate(axs.flatten()):
    ax.set_xlabel("Wartość zmiennych losowych")
    ax.set_ylabel("Gęstość prawdopodobieństwa")
```

#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.

#### Rozkład normalny
Wartość oczekiwana wpływa na to gdzie zlokalizowany jest środek rozkładu - większa wartość 𝜇 sprawia, że cała funkcja przesuwa się w prawo, a zmniejszenie 𝜇 przesuwa rozkład w lewo. Parametr 𝜎 określa jak "rozciągnięty" jest rozkład wzdłuż osi x, zmniejszenie wartości 𝜎 sprawia, że staje się on węższy i wyższy, czyli wartości skupiają się bliżej wartości oczekiwanej, natomiast następstwem zwiększenia wartości 𝜎 jest zpłaszczenie i rozciągnięcie funkcji wzdłuż osi x.
    
#### Rozkład t-Studenta
Liczba stopni swobody 𝑑𝑓 w rozkładzie t-Studenta wpływa na wysokość rozkładu i wygląd "ogonów" funkcji, zwiększenie liczby stopni swobody sprawia, że zmniejssza się wysokość a ogony funkcji są "cięższe", czyli zwiększa się prawdopodobieństwo występowania wartości bardziej oddalonych od wartości oczekiwanej.

#### Rozkład Chi-kwadrat
Rozkład jest prawostronnie asymetryczny - liczba stopni swobody 𝑑𝑓 w rozkładzie chi-kwadrat wpływa na symetryczność krzywej, dla niskich wartości 𝑑𝑓 krzywa jest mocno asymetryczna i skośna, zwiększanie liczby stopni swobody powoduje, że rozkład zbliża się wyglądem do kształtu dzwonu. Większa liczba stopni swobody powoduje także, że wartości są mocnej skupione wokół wartości oczekiwanej. Ze względu na jego specyfikę nigdy nie przyjmuje wartości ujemnych.

#### Rozkład F Snedecora
Rozkład jest asymetryczny i prawostronnie skośnny. Gdy oba stopnie swobody rosną zaczyna przypominać rozkład normalny. Gdy wzrasta jeden ze stopni swobody rozkład przesuwa się w prawo. Gdy pierwszy stopień swobody ma małą wartość - 1 lub 2, wysokość rozkładu drastycznie  rośnie. Wysokość również wzrasta gdy zwiększają się na raz oba stopnie swobody i są sobie równe. Rozkład nie przyjmuje wartości ujemnych.

### Zadanie 2: Praca z rozkładami teoretycznymi [10 pkt]

#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
df = 10
p1 = sp.stats.t.cdf(-1.5, df)
x = np.linspace(sp.stats.t.ppf(0.0001, df), sp.stats.t.ppf(0.9999, df), 1000)
y = sp.stats.t.pdf(x, df)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'r-', label='t(10) Density')

pole_x = np.linspace(sp.stats.t.ppf(0.0001, df), -1.5, 100)
pole_y = sp.stats.t.pdf(pole_x, df)
plt.fill_between(pole_x, 0, pole_y, where=(pole_x <= -1.5), color='pink', alpha=0.8)

plt.xlim(sp.stats.t.ppf(0.0001, df), sp.stats.t.ppf(0.9999, df))
plt.ylim(0, 0.4)
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(X < -1.5) = {p1:.4f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# p2

p2 = 1 - sp.stats.f.cdf(2, 5, 50)

x = np.linspace(sp.stats.f.ppf(0.0001, 5, 50), sp.stats.f.ppf(0.9999, 5, 50), 1000)
y = sp.stats.f.pdf(x, 5, 50)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'r-', label='F(5, 50) Density')

pole_x = np.linspace(2, sp.stats.f.ppf(0.9999, 5, 50), 100)
pole_y = sp.stats.f.pdf(pole_x, 5, 50)
plt.fill_between(pole_x, 0, pole_y, where=(pole_x >= 2), color='pink', alpha=0.8)

plt.xlim(sp.stats.f.ppf(0.0001, 5, 50), sp.stats.f.ppf(0.9999, 5, 50))
plt.ylim(0, 1)
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(X > 2) = {p2:.4f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#p3
df = 4
p3 = sp.stats.chi2.cdf(10, df) - sp.stats.chi2.cdf(3, df)

x = np.linspace(sp.stats.chi2.ppf(0.0001, df), sp.stats.chi2.ppf(0.9999, df), 1000)
y = sp.stats.chi2.pdf(x, df)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'r-', label='chi2(4) Density')

pole_x = np.linspace(3, 10, 100)
pole_y = sp.stats.chi2.pdf(pole_x, df)
plt.fill_between(pole_x, 0, pole_y, where=((pole_x <= 10) & (pole_x >= 3)), color='pink', alpha=0.8)

plt.xlim(sp.stats.chi2.ppf(0.0001, df), sp.stats.chi2.ppf(0.9999, df))
plt.ylim(0, 0.3)
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(3 <= X <= 10) = {p1:.4f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#p4
p4 = 1 - sp.stats.norm.cdf(1.5, -0.5, 1.5) + sp.stats.norm.cdf(-2.75, -0.5, 1.5)

x = np.linspace(-5, 5, 1000)
y = sp.stats.norm.pdf(x, -0.5, 1.5)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'r-', label='N(-0.5, 1.5) Density')

pole_x = np.linspace(-5, 5, 100)
pole_y = sp.stats.norm.pdf(pole_x, -0.5, 1.5)
plt.fill_between(pole_x, 0, pole_y, where=((pole_x <= -2.75) | (pole_x >= 1.5)), color='pink', alpha=0.8)

plt.xlim(-5, 5)
plt.ylim(0, 0.3)
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(X < -2.75 lub X > 1.5) = {p4:.4f}')
plt.legend()
plt.grid(True)
plt.show()
print(p1)
print(p2)
print(p3)
print(p4)
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

Rozwiąż i zapisz do zmiennych o podanych nazwach rozwiązania następujących równań:
- `s1` - $P(X < s) = 0.25$, gdzie X to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 5 stopniach swobody - $\chi^2(5)$,
- `s2` - $P(X > s) = 0.15$, gdzie X to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 20 stopniach swobody - $F(5, 20)$,
- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- oznacz krawędź obszaru, na której znajduje się wyznaczona wartość $s$ linią w wyróżniającym się kolorze,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość zmiennej $X$ podaną do 4 miejsca po przecinku (np. P(X < s.ssss) = 0.25, gdzie s.ssss to obliczona wartość).

```python
# s1
s1 = sp.stats.chi2.ppf(0.25, 5)
x = np.linspace(sp.stats.chi2.ppf(0.0001, 5), sp.stats.chi2.ppf(0.9999, 5), 1000)
y = sp.stats.chi2.pdf(x, 5)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'red', label='Rozkład chi2(5)')

plt.fill_between(x, y, where=(x <= s1), color='pink', alpha=0.8)

plt.axvline(x=s1, color='k', linestyle='--', label=f's = {s1:.4f}')

plt.xlim(sp.stats.chi2.ppf(0.0001, 5), sp.stats.chi2.ppf(0.9999, 5))
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(X < {s1:.4f}) = 0.25')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# s2
s2 = sp.stats.f.ppf(0.85, 5, 20)

x = np.linspace(sp.stats.f.ppf(0.0001, 5, 20), sp.stats.f.ppf(0.9999, 5, 20), 1000)
y = sp.stats.f.pdf(x, 5, 20)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'red', label='Rozkład F(5,20)')

plt.fill_between(x, y, where=(x >= s2), color='pink', alpha=0.8)

plt.axvline(x=s2, color='k', linestyle='--', label=f's = {s1:.4f}')

plt.xlim(sp.stats.f.ppf(0.0001, 5, 20), sp.stats.f.ppf(0.9999, 5, 20))
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(X > {s2:.4f}) = 0.15')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# s3
s = sp.stats.norm.cdf(-1.25, -0.4, 0.75)
s3 = sp.stats.norm.ppf(0.35+s, -0.4, 0.75)

x = np.linspace(sp.stats.norm.ppf(0.0001, -0.4, 0.75), sp.stats.norm.ppf(0.9999, -0.4, 0.75), 1000)
y = sp.stats.norm.pdf(x, -0.4, 0.75)

plt.figure(figsize=(10, 6), facecolor='white')
plt.plot(x, y, 'red', label='Rozkład N(-0.4, 0.75)')

plt.fill_between(x, y, where=((x > -1.25) & (x < s3)), color='pink', alpha=0.8)

plt.axvline(x=s3, color='k', linestyle='--', label=f's = {s3:.4f}')
plt.axvline(x=-1.25, color='k', linestyle='--', label=f'x = 1.25')

plt.xlim(sp.stats.norm.ppf(0.0001, -0.4, 0.75), sp.stats.norm.ppf(0.9999, -0.4, 0.75))
plt.xlabel('X')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'P(-1.25 < X < {s3:.4f}) = 0.35')
plt.legend()
plt.grid(True)
plt.show()
```
